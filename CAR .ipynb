```python
from tkinter import *
import joblib
import pandas as pd

def show_entry_fields():
    try:
        p1 = float(e1.get())
        p2 = float(e2.get())
        p3 = float(e3.get())
        p4 = float(e4.get())
        p5 = float(e5.get())
        p6 = float(e6.get())
        p7 = float(e7.get())
        p8 = float(e8.get())

        model = joblib.load('car_price_predictor')
        data_new = pd.DataFrame({
            'Year': p1,
            'Present_Price': p2,
            'Kms_Driven': p3,
            'Fuel_Type': p4,
            'Seller_Type': p5,
            'Transmission': p6,
            'Owner': p7,
            'Age': p8
        }, index=[0])

        result = model.predict(data_new)
        result_label.config(text=f"Predicted Price: â‚¹ {round(result[0], 2)}", fg="green", font=("Arial", 12, "bold"))
        print("Car Purchase Amount", result[0])
    except Exception as e:
        result_label.config(text=f"Error: {e}", fg="red", font=("Arial", 10, "bold"))

master = Tk()
master.title("Car Price Prediction")
master.configure(bg="#f2f2f2") 


Label(master, text="Car Price Prediction Using Machine Learning", bg="black", fg="white",
      font=("Helvetica", 14, "bold"), pady=10).grid(row=0, column=0, columnspan=2, sticky="ew")

# Input Labels
labels = ["Year", "Present_Price", "Kms_Driven", "Fuel_Type", "Seller_Type", "Transmission", "Owner", "Age"]
entries = []

for i, label_text in enumerate(labels):
    Label(master, text=label_text, bg="#f2f2f2", font=("Arial", 10)).grid(row=i+1, column=0, padx=10, pady=5, sticky=W)
    entry = Entry(master)
    entry.grid(row=i+1, column=1, padx=10, pady=5)
    entries.append(entry)

e1, e2, e3, e4, e5, e6, e7, e8 = entries

# Predict button
Button(master, text='Predict', command=show_entry_fields, bg="#4CAF50", fg="white",
       font=("Arial", 10, "bold"), padx=10).grid(row=10, column=0, columnspan=2, pady=10)


result_label = Label(master, text="", bg="#f2f2f2")
result_label.grid(row=11, column=0, columnspan=2, pady=5)

mainloop()
```
